# Ejercicio 1: Conversión de Temperatura

A continuación se presenta la solución planteada para este problema:

```c
# include <stdio.h>

float conversion_farenheit(float celsius){
    return ((9.0 / 5.0) * celsius) + 32.0;
}

float conversion_kelvin(float celsius){
    return (celsius + 273.15);
}

int main(){

    float temperatura_celsius, temperatura_farenheit, temperatura_kelvin;

    printf("Por favor ingrese una temperatura en Grados Celsius:\n");
    scanf("%f", &temperatura_celsius);

    temperatura_farenheit = conversion_farenheit(temperatura_celsius);
    temperatura_kelvin = conversion_kelvin(temperatura_celsius);

    printf("La temperatura ingresada fue de %.2f grados celsius. \nEn Grados Farenheit es: %.2f\nEn grados Kelvin es: %.2f",
        temperatura_celsius, temperatura_farenheit, temperatura_kelvin);

    return 0;
}
```

# Ejercicio 2: Cálculo del Área de un Triángulo

A continuación se presenta la solución planteada para este problema:

```c
#include <stdio.h>
#include <math.h>


int main(){

    float lado_a, lado_b, lado_c, semip_p, area;
    int datos_correctos= 0;


    while(datos_correctos == 0){
        
        printf("Ingrese el valor del lado A(cm): ");
        scanf("%f", &lado_a);

        printf("Ingrese el valor del lado B(cm): ");
        scanf("%f", &lado_b);

        printf("Ingrese el valor del lado C(cm): ");
        scanf("%f", &lado_c);

        //Comprobación de la desigualdad triangular
        if (lado_a + lado_b < lado_c || lado_b + lado_c < lado_a || lado_c + lado_a < lado_b){
            printf("Los lados no cumplen con la desigualdad triangular\n");
            printf("Vuelva a ingresar los datos\n");
            datos_correctos = 0;
        }
        // Los lados de un triangulo deben ser mayores que 0
        else if (lado_a <= 0 || lado_b <= 0 || lado_c <= 0){
            printf("\nLos lados deben ser mayores a cero\n");
            printf("Vuelva a ingresar los datos\n");
            datos_correctos = 0;
        }
        else{
            datos_correctos = 1;
        }
        
    }

    // Calculo del semiperimetro
    semip_p = (lado_a + lado_b + lado_c) / 2.0;

    // Calculo del area
    area = sqrt(semip_p * (semip_p - lado_a)*(semip_p - lado_b)*(semip_p - lado_c));
    
    printf("El area del triangulo es: %.2f cm^2\n", area);
    

    return 0;
}
```

A continuación se presenta la solución planteada para este problema:

```c
# include <stdio.h>

int suma_numeros(int num_max){
    int suma_numeros = 0;
    int num = 0;
    while (num <= num_max){
        suma_numeros = suma_numeros + num;
        num ++;
    }

    return suma_numeros;
}

int main(){

    int cantidad_num_naturales, suma;

    printf("\nEste Programa realiza la suma de los primeros 'n' numeros naturales\n\nPor favor ingrese 'n': ");
    scanf("%d", &cantidad_num_naturales);
    printf("\n!!!ADVERTENCIA!!!\nSi ingreso un numero que no es entero el programa tomara la parte entera y botara los decimales");

    suma = suma_numeros(cantidad_num_naturales);
    printf("\n\nLa suma de los primeros %d numeros naturales es: %d", cantidad_num_naturales, suma);

    return 0;
}
```

# Ejercicio 4: Determincación de un Número Primo

A continuación se presenta la solución planteada para este problema:
```c
# include <stdio.h>
# include <math.h>


char *es_primo(int numero){
    char *resultado;
    if (numero <= 1){
        resultado = "no es";
    }
    else if (numero == 2 || numero == 3){
        resultado = "es";
    }

    else if (numero % 2 == 0 || numero % 3 == 0){
        resultado = "no es";
    }
    else{
        float num_max = sqrt(numero);
        resultado = "es";
        for (int divisor = 2; divisor <= num_max; divisor ++){  
            if (numero % divisor == 0){
                resultado = "no es";
                break;
            }
        }
    }

    return resultado;
}

int main(){

    int numero;
    char *resultado;

    printf("\nIngrese un numero NATURAL para determinar si es un numero primo o no: ");
    scanf("%d", &numero);

    resultado = es_primo(numero);
    printf("\nEl numero ingresado %s un numero primo\n\n", resultado);
    return 0;
}
```

* definición de la estructura de datos
* función `insertarFinal`
* función `mostrarLista`
* función `eliminarNodo`



---

## Definición de la estructura de datos

```c
struct Nodo {
    int dato;
    struct Nodo* siguiente;
};
```

Aquí declaramos `struct Nodo` con un `dato` donde estára el valor de cada nodo, en este caso particular un número entero y un puntero `siguiente` que apunta al próximo elemento de la lista. Esto mejora el manejo de la memoria ya que no requiere un bloque de espacio continuo en memoria a diferencia de un array por lo que además de la posibilidad de agregar dinámicamente, es menos probable que nos de un error por overflow a la hora de usar `malloc`.

---

## Función `insertarFinal`

```c
void insertarFinal(struct Nodo** cabeza, int valor) {
    struct Nodo* nuevoNodo = malloc(sizeof(struct Nodo));
    if (nuevoNodo == NULL) {
        perror("malloc");
        exit(EXIT_FAILURE);
    }
    nuevoNodo->dato = valor;
    nuevoNodo->siguiente = NULL;

    if (*cabeza == NULL) {
        *cabeza = nuevoNodo;
    } else {
        struct Nodo* temp = *cabeza;
        while (temp->siguiente != NULL) {
            temp = temp->siguiente;
        }
        temp->siguiente = nuevoNodo;
    }
}
```
En esta función recibimos como parámetro un puntero doble a la cabeza de la lista (`struct Nodo** cabeza`) y el valor a insertar. El puntero doble es la clave en este punto, para esta parte se usó la **ayuda de la IA**, ya que al realizar el código solo había usado un puntero, por lo que no modificaba la dirección de forma global si no local, este fue un error simple de corregir pero fundamental para comprender el acceso a los punteros y las diferentes formas de modificar los mismos, ahora bien, este puntero se usa para poder modificar la dirección de la cabeza dentro de la propia función: cuando la lista está vacía, basta con definir `*cabeza = nuevoNodo` para que este nodo sea el primero. Reservamos memoria en el heap con `malloc`, comprobamos que no devuelva `NULL` y asignamos el valor de `dato`. Si la lista no estaba vacía, usamos un puntero temporal `temp` que recorre de nodo en nodo hasta que se encuentra que `siguiente` es `NULL`; en ese punto enlazamos el nuevo nodo. La variable `temp` permite avanzar sin alterar la referencia original de la cabeza.

---

## Función `mostrarLista`

```c
void mostrarLista(struct Nodo* cabeza) {
    if (cabeza == NULL) {
        printf("Lista vacía\n");
        return;
    }
    struct Nodo* temp = cabeza;
    while (temp != NULL) {
        printf("%d -> ", temp->dato);
        temp = temp->siguiente;
    }
    printf("NULL\n");
}
```
Para visualizar el contenido de la lista, usamos un puntero  `temp` que  copia la dirección de la cabeza. Sin modificar `cabeza`, en cada ciclo `temp` avanza nodo a nodo gracias a `temp = temp->siguiente`, imprimiendo en cada paso el valor almacenado. El ciclo se detiene cuando `temp` se vuelve `NULL`, momento en el cual la cadena de impresión se acaba imprimiendo "NULL", indicando que el último elemento de la lista apunta al nulo. Para evitar errores, `cabeza == NULL` evita accesos inválidos en listas vacías.

---

## Función `eliminarNodo`

```c
void eliminarNodo(struct Nodo** cabeza, int valor) {
    if (*cabeza == NULL) {
        printf("Lista vacía\n");
        return;
    }
    if ((*cabeza)->dato == valor) {
        struct Nodo* aux = *cabeza;
        *cabeza = aux->siguiente;
        free(aux);
        return;
    }
    struct Nodo* temp = *cabeza;
    while (temp->siguiente != NULL && temp->siguiente->dato != valor) {
        temp = temp->siguiente;
    }
    if (temp->siguiente == NULL) {
        printf("Valor no encontrado\n");
    } else {
        struct Nodo* nodoAEliminar = temp->siguiente;
        temp->siguiente = nodoAEliminar->siguiente;
        free(nodoAEliminar);
    }
}
```
Eliminar un nodo requiere, de nuevo, un puntero doble para  el caso de borrar el primer elemento *(en este caso se uso la IA para lo mismo que en la función de agregar)*

Eliminar un nodo en una lista enlazada requiere, nuevamente, el uso de un puntero doble, lo cual es fundamental especialmente cuando el nodo a eliminar es el primero de la lista. En este caso, se hace algo similar a la función de inserción: si `(*cabeza)->dato` coincide con el valor a eliminar, primero se guarda la dirección del nodo actual en un puntero auxiliar, luego se actualiza el puntero a la cabeza con `*cabeza = aux->siguiente`, y finalmente se libera la memoria de `aux` usando `free(aux)`. Esto permite eliminar el primer nodo sin perder la referencia al resto de la lista.

En el caso de que el nodo a eliminar se encuentre en una posición intermedia o final, no basta con recorrer la lista hasta encontrar el nodo con el dato deseado; es necesario detenerse en el nodo anterior, es decir, pararnos en el nodo cuyo `siguiente` apunta al nodo a eliminar. Para esto, se puede usa un puntero `temp` que recorra la lista hasta cumplir esta condición. Cuando lo encuentra, se almacena la dirección del nodo a eliminar en otro puntero, en este caso `nodoAEliminar`, se ajusta el enlace con `temp->siguiente = nodoAEliminar->siguiente` para saltarse el nodo que se quería eliminar, y finalmente se libera su memoria con `free(nodoAEliminar)`.

# Ejercicio 8: Ordenamiento con Burbuja

### Proceso para resolver el problema

En una clase de HPC mientras aprendíamos el funcionamiento de los punteros se nos pidió hacer un código implementando el bubble sort, y durante la clase se discutió la solución del mismo, por lo que gran parte del código sale de lo que se realizó en esta clase.

Para abordar la solución de este ejercicio, se ha decidido estructurar la explicación en tres partes principales que reflejan claramente la lógica detrás del algoritmo de ordenamiento propuesto:  
* Inicialización de variables y punteros  
* Implementación del algoritmo de ordenamiento tipo burbuja  
* Presentación del resultado (impresión de los valores ordenados)

A continuación se presenta el código completo:

```c
#include <stdio.h>

int main()
{
    int numeros[] = {4, 1, 2, 3, 5};
    int *dir_i, *dir_j;
    int i, j, aux;
    size_t size = sizeof(numeros) / sizeof(numeros[0]);
    
    dir_i = numeros;
    dir_j = numeros;
    
    for(i = 0; i < size; i++){
        for(j = 0; j < size - 1; j++){
            if(*dir_j > *(dir_j + 1)){
                aux = *dir_j;
                *dir_j = *(dir_j+1);
                *(dir_j+1) = aux;
            }
            dir_j = numeros + j;
        }
    }
    
    dir_i = numeros;
    
    for(i=0; i < size; i++){
        printf("%d\n" , *dir_i);
        dir_i = dir_i + 1;
    }
    
    return 0;
}
```

---

## Inicialización de variables y punteros

En esta primera sección del código, se define un arreglo de enteros con algunos valores desordenados:

```c
int numeros[] = {4, 1, 2, 3, 5};
```

A continuación, se declaran dos punteros, `dir_i` y `dir_j`, con los que podemos recorrer el arreglo a través de las direcciones de memoria en lugar de usar los índices directamente. También se define una variable auxiliar `aux`, para hacer los intercambios entre los elementos:

```c
int *dir_i, *dir_j;
int i, j, aux;
```

Para determinar cuántos elementos tiene el arreglo, se utiliza la expresión `sizeof(numeros) / sizeof(numeros[0])`, que nos permite conocer cuántos bytes ocupa el arreglo y dividirlo por el tamaño de uno de sus elementos. Este valor se guarda en la variable `size`:

```c
size_t size = sizeof(numeros) / sizeof(numeros[0]);
```

Finalmente, se inicializan los punteros `dir_i` y `dir_j` apuntando a la primera dirección del arreglo:

```c
dir_i = numeros;
dir_j = numeros;
```

---


Esta es la parte del código en la cual se usa el algoritmo de ordenamiento burbuja, para ello, se utilizan dos ciclos `for` anidados. El primero controla el número de pasadas necesarias y el segundo compara pares consecutivos de elementos:

```c
for(i = 0; i < size; i++){
    for(j = 0; j < size - 1; j++){
        if(*dir_j > *(dir_j + 1)){
            aux = *dir_j;
            *dir_j = *(dir_j+1);
            *(dir_j+1) = aux;
        }
        dir_j = numeros + j;
    }
}
```

Aquí el puntero `dir_j` se está utilizando para comparar valores directamente a través de la desreferencia. Es decir, se compara el valor que está en la dirección `dir_j` con el que está una posición más adelante (`dir_j + 1`). Si el primero es mayor, se hace un intercambio utilizando la variable `aux`. Este proceso se repite en cada pasada hasta que el arreglo queda completamente ordenado.



---

## Presentación del resultado

Una vez que el arreglo ha sido ordenado, se vuelve a utilizar el puntero `dir_i` para recorrer y mostrar los valores ordenados uno por uno:

```c
dir_i = numeros;

for(i=0; i < size; i++){
    printf("%d\n" , *dir_i);
    dir_i = dir_i + 1;
}
```

Aquí `dir_i` se inicializa de nuevo apuntando al primer elemento del arreglo, y a través de la aritmética de punteros (`dir_i + 1`) se avanza posición por posición, imprimiendo el valor contenido en cada una.

### Implementación de pila con listas enlazadas en C

Así como se mencionó en la implementación de la lista enlazada simple, este ejercicio también hace uso de una estructura dinámica que permite gestionar memoria de forma eficiente. En este caso, abordamos la creación de una **pila** (estructura LIFO: Last In, First Out) utilizando listas enlazadas. Vamos a explicar el código dividiéndolo en secciones clave:

---

## Definición de la estructura de datos

```c
typedef struct Nodo {
    int dato;                 // Valor almacenado en el nodo
    struct Nodo* siguiente;   // Puntero al siguiente nodo en la pila
} Nodo;
```

De manera similar a la lista enlazada simple, se declara un `struct Nodo` que almacena un valor entero y un puntero al siguiente nodo. En este contexto, los nodos se apilan uno sobre otro, donde cada nuevo nodo apunta al nodo que estaba en el tope.

---

## Función `push` (apilar elementos)

```c
void push(Nodo** tope, int valor) {
    Nodo* nuevoNodo = malloc(sizeof(Nodo));
    if (nuevoNodo == NULL) {
        fprintf(stderr, "Error: malloc falló\n");
        exit(EXIT_FAILURE);
    }
    nuevoNodo->dato = valor;
    nuevoNodo->siguiente = *tope;
    *tope = nuevoNodo;
}
```

Como se vio previamente en la función `insertarFinal`, el uso de un **puntero doble** permite modificar la dirección de forma global dentro de la función. En este caso, `*tope` apunta al nodo superior de la pila. Con `malloc` se reserva memoria para el nuevo nodo, se le asigna el valor y se enlaza como el nuevo tope.

---

## Función `pop` (desapilar elementos)

```c
int pop(Nodo** tope) {
    if (*tope == NULL) {
        fprintf(stderr, "Error: pila vacía\n");
        return -1;
    }
    Nodo* temp = *tope;
    int valor = temp->dato;
    *tope = temp->siguiente;
    free(temp);
    return valor;
}
```

Así como en la función `eliminarNodo`, se hace uso del puntero doble para modificar la cabeza (en este caso el tope). Se verifica primero si la pila está vacía. Si no lo está, se extrae el valor del nodo superior, se actualiza el puntero `tope` y se libera la memoria ocupada por el nodo anterior.

---

## Función `mostrar` (recorrer y mostrar la pila)

```c
void mostrar(Nodo* tope) {
    if (tope == NULL) {
        printf("Pila vacía\n");
        return;
    }
    printf("Contenido de la pila:\n");
    for (Nodo* actual = tope; actual != NULL; actual = actual->siguiente) {
        printf("%d\n", actual->dato);
    }
}
```

Similar a la función `mostrarLista`, se usa un puntero auxiliar (`actual`) para recorrer la pila desde el tope hasta que se llega al final (donde el siguiente es `NULL`). No se modifica el puntero original al tope.

---

## Función `liberarPila` (liberar toda la memoria)

```c
void liberarPila(Nodo** tope) {
    while (*tope != NULL) {
        Nodo* temp = *tope;
        *tope = temp->siguiente;
        free(temp);
    }
}
```

Así como en la lista enlazada liberamos nodo por nodo en `eliminarNodo`, aquí se hace un recorrido desde el tope liberando la memoria asignada con `malloc`.

---

## Función principal `main` (ejemplo de uso)

```c
int main() {
    Nodo* pila = NULL;
    push(&pila, 5);
    push(&pila, 10);
    push(&pila, 15);
    mostrar(pila);
    printf("Desapilado: %d\n", pop(&pila));
    mostrar(pila);
    liberarPila(&pila);
    return 0;
}
```

Se crea una pila vacía (`Nodo* pila = NULL`) y se agregan elementos con `push`. Luego, se elimina un elemento con `pop` y se muestra el contenido antes y después. Finalmente, se libera toda la memoria.

---

## Punteros y espacio en memoria

Como se ha mencionado también en el ejercicio 7, el **puntero simple** se usa para recorrer sin modificar la función principal, mientras que el **puntero doble** es esencial para modificar la dirección original desde una función. Y evidenciamos que para toda operación de asignación de memoria con `malloc` debe usarse `free` cuando ya no se necesite esa dirección en memoria, todo esto para gestionar correctamente el uso de memoria dinámica en C.
