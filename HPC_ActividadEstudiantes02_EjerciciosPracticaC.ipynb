#### **Introducción**
El área de un triángulo cuyos lados son $a$, $b$ y $c$ se puede calcular mediante la **fórmula de Herón**:

1. Se calcula el **semiperímetro**:

   $$
   s = \frac{a + b + c}{2}
   $$

2. Se usa la siguiente fórmula para el área:

   $$
   A = \sqrt{s(s - a)(s - b)(s - c)}
   $$

Para que los lados formen un triángulo válido, deben cumplir la desigualdad triangular:

$$
a + b > c, \quad a + c > b, \quad b + c > a
$$

#### **Problema**

### **4. Determinación de un Número Primo (Uso de Funciones)**
#### **Introducción**
Un número primo es un número natural mayor que 1 que solo tiene **dos divisores**: `1` y él mismo. Ejemplos:

- **Primos**: 2, 3, 5, 7, 11, 13, 17, ...
- **No primos**: 4 (divisores: 1, 2, 4), 6 (divisores: 1, 2, 3, 6).

El número `1` **no es primo**, ya que solo tiene un divisor (él mismo).  
El número `2` **es el único número primo par**, ya que todos los números pares mayores que `2` son divisibles por `2` y, por lo tanto, no pueden ser primos.


#### **Problema**
Escriba un programa que determine si un número ingresado es primo. Para ello:
1. Implemente una función `esPrimo(int n)`.
2. Llame a la función desde `main()` y muestre el resultado.

### **Estrategia Algorítmica**
Para verificar si un número $ n $ es primo, no es necesario probar todos los números desde `2` hasta `n-1`. Podemos optimizar el algoritmo observando lo siguiente:

1. Si $ n $ es **menor o igual a 1**, no es primo.
2. Si $ n $ es **igual a 2 o 3**, es primo.
3. Si $ n $ es **par o divisible por 3**, no es primo.
4. Solo necesitamos verificar **divisores hasta $ \sqrt{n} $**, porque si $ n $ tiene un divisor mayor que $ \sqrt{n} $, el otro factor debe ser menor que $ \sqrt{n} $.
5. Podemos reducir el número de pruebas aún más, verificando solo los números impares mayores que `2` (ya que los pares no pueden ser primos).

#### **Objetivo**
- Introducir el concepto de **funciones** en C.
- Manejo de ciclos y operadores de comparación.

### **6. Suma de Matrices (Uso de Arreglos Bidimensionales)**
#### **Introducción**
Las **matrices** son estructuras fundamentales en computación, utilizadas en gráficos, procesamiento de imágenes y resolución de ecuaciones.

La **suma de matrices** es una operación matemática en la que se suman los elementos correspondientes de dos matrices del mismo tamaño.

#### **Definición Matemática**
Dadas dos matrices $ A $ y $ B $ de tamaño $ m \times n $, la suma $ C $ se define como:

$$
C[i][j] = A[i][j] + B[i][j]
$$

donde:
- $ A[i][j] $ es el elemento de la fila $ i $ y columna $ j $ en la matriz $ A $.
- $ B[i][j] $ es el elemento correspondiente en la matriz $ B $.
- $ C[i][j] $ es el resultado de la suma en la posición $ (i, j) $.

### **Pseudocódigo**
```plaintext
Inicio
    Leer m, n  // Número de filas y columnas de las matrices
    
    Definir matriz A[m][n]
    Definir matriz B[m][n]
    Definir matriz C[m][n]
    
    // Leer la matriz A
    Para i ← 0 hasta m - 1 hacer
        Para j ← 0 hasta n - 1 hacer
            Leer A[i][j]
    
    // Leer la matriz B
    Para i ← 0 hasta m - 1 hacer
        Para j ← 0 hasta n - 1 hacer
            Leer B[i][j]
    
    // Sumar A y B, almacenando en C
    Para i ← 0 hasta m - 1 hacer
        Para j ← 0 hasta n - 1 hacer
            C[i][j] ← A[i][j] + B[i][j]
    
    // Imprimir la matriz C
    Para i ← 0 hasta m - 1 hacer
        Para j ← 0 hasta n - 1 hacer
            Escribir C[i][j]

Fin
```
#### **Problema**
Escriba un programa en C que:
1. lea desde archivos de datos cada una de las dos matrices $ A $ y $ B $ de tamaño $ m \times n $.
2. Calcule la suma $ C = A + B $.
3. Muestre la matriz resultante.

### **7. Implementación de una Lista Enlazada Simple (Uso de Structs y Punteros)**

#### **Introducción**
Las **listas enlazadas** son estructuras de datos dinámicas que permiten insertar y eliminar elementos sin necesidad de reorganizar la memoria.

#### **Concepto Matemático**

Una **lista enlazada simple** se compone de **nodos**, donde cada nodo contiene:
1. **Un valor** (dato almacenado).
2. **Un puntero al siguiente nodo** en la lista.

Visualmente, una lista enlazada con los valores `[5, 10, 15]` se representa así:

```
[5 | *] → [10 | *] → [15 | NULL]
```
Donde `*` representa el puntero al siguiente nodo y `NULL` indica el final de la lista.

#### **Operaciones en una Lista Enlazada Simple**
Las principales operaciones en una lista enlazada son:
1. **Inserción al final**: Agregar un nodo con un nuevo valor.
2. **Mostrar la lista**: Recorrer y mostrar los elementos de la lista.
3. **Eliminar un nodo**: Borrar un nodo dado su valor.

##### **Pseudocódigo**
- **Definir la estructura del nodo**
```plaintext
Definir estructura Nodo:
    Dato → entero
    Siguiente → Puntero a Nodo
```

- **Insertar un nodo al final**
```plaintext
Función insertarFinal(Lista, valor):
    Crear nuevoNodo con valor
    Si Lista es NULL entonces
        Lista ← nuevoNodo
    Sino
        temp ← Lista
        Mientras temp.Siguiente ≠ NULL hacer
            temp ← temp.Siguiente
        FinMientras
        temp.Siguiente ← nuevoNodo
    FinSi
FinFunción
```

- **Mostrar la lista enlazada**
```plaintext
Función mostrarLista(Lista):
    Si Lista es NULL entonces
        Escribir "Lista vacía"
    Sino
        temp ← Lista
        Mientras temp ≠ NULL hacer
            Escribir temp.Dato
            temp ← temp.Siguiente
        FinMientras
    FinSi
FinFunción
```

- **Eliminar un nodo dado su valor**
```plaintext
Función eliminarNodo(Lista, valor):
    Si Lista es NULL entonces
        Escribir "Lista vacía"
        Retornar

    Si Lista.Dato = valor entonces
        Lista ← Lista.Siguiente
        Retornar

    temp ← Lista
    Mientras temp.Siguiente ≠ NULL y temp.Siguiente.Dato ≠ valor hacer
        temp ← temp.Siguiente
    FinMientras

    Si temp.Siguiente = NULL entonces
        Escribir "Valor no encontrado"
    Sino
        temp.Siguiente ← temp.Siguiente.Siguiente
    FinSi
FinFunción
```

#### **Explicación del Pseudocódigo**
1. **Definir un nodo**: Se usa una estructura con dos elementos: el valor (`Dato`) y un puntero al siguiente nodo (`Siguiente`).
2. **Insertar un nodo al final**:
   - Si la lista está vacía, el nuevo nodo se convierte en la lista.
   - Si la lista tiene elementos, se recorre hasta el último nodo y se enlaza el nuevo nodo.
3. **Mostrar la lista**:
   - Se recorre desde el primer nodo hasta `NULL`, imprimiendo cada dato.
4. **Eliminar un nodo**:
   - Si el nodo a eliminar es el primero, la lista se actualiza al siguiente nodo.
   - Si el nodo está en el medio o final, se busca y se ajusta el puntero del nodo anterior.

#### **Ejemplo Paso a Paso**
1. **Insertar 5, 10, 15 en una lista vacía**:
```
Lista: NULL
Insertar(5) → [5 | NULL]
Insertar(10) → [5 | *] → [10 | NULL]
Insertar(15) → [5 | *] → [10 | *] → [15 | NULL]
```
2. **Eliminar 10**:
```
Buscar nodo con 10
Modificar puntero: [5 | *] → [15 | NULL]
Lista después de eliminar: [5 | *] → [15 | NULL]
```

#### **Complejidad Temporal**
| Operación       | Complejidad |
|----------------|------------|
| **Insertar al final** | O(n) |
| **Mostrar elementos** | O(n) |
| **Eliminar nodo** | O(n) |

Donde `n` es el número de elementos en la lista.

#### **Resumen**
- **Concepto**: La lista enlazada es una estructura donde cada nodo almacena un dato y un puntero al siguiente nodo.
- **Estrategia**: Manipular punteros para agregar, recorrer y eliminar nodos sin mover grandes bloques de memoria.
- **Aplicaciones**: Manejo dinámico de memoria, estructuras como pilas y colas, almacenamiento flexible.

### **8. Ordenamiento con Burbuja (Uso de Punteros)**
#### **Introducción**
El **ordenamiento de burbuja** es un algoritmo de clasificación simple que **compara y cambia de posición** elementos adyacentes hasta que la lista esté ordenada. Su nombre se debe a que los valores más grandes van "flotando" hasta su posición correcta, como burbujas en un líquido.

#### **Definición Matemática**
Dado un arreglo de `n` elementos:
$$
A = [a_1, a_2, a_3, ..., a_n]
$$
El **ordenamiento de burbuja** consiste en realizar múltiples pasadas sobre el arreglo, comparando e **intercambiando elementos adyacentes** si están en el orden incorrecto. Se repite el proceso hasta que **no haya más intercambios**.

Para cada iteración `i`, se compara cada par de elementos consecutivos `A[j]` y `A[j+1]`:
- Si **`A[j] > A[j+1]`**, los elementos se intercambian.
- Al final de la primera pasada, el elemento más grande estará en la última posición.
- Se repite el proceso ignorando la última posición en cada iteración.

#### **Ejemplo Numérico**
Dado el arreglo:

```
[5, 3, 8, 1, 2]
```
Se ordena paso a paso:

| Iteración | Arreglo Intermedio  | Intercambios |
|-----------|---------------------|--------------|
| **1**     | [3, 5, 1, 2, 8]     | (5↔3), (8↔1), (8↔2) |
| **2**     | [3, 1, 2, 5, 8]     | (5↔1), (5↔2) |
| **3**     | [1, 3, 2, 5, 8]     | (3↔1), (3↔2) |
| **4**     | [1, 2, 3, 5, 8]     | (2↔3) |

El arreglo ahora está ordenado en **O(n²) pasos** en el peor caso.

#### **Algoritmo para el Ordenamiento de Burbuja**
El **ordenamiento de burbuja** puede implementarse de dos maneras:
1. **Usando índices de arreglos (`A[i]`)**.
2. **Usando punteros (`*(ptr + i)`)**, lo que mejora la eficiencia en C.

##### **Estrategia con punteros**
1. Definir un **puntero** al arreglo.
2. Usar dos bucles anidados:
   - El **bucle externo** controla las pasadas sobre el arreglo.
   - El **bucle interno** compara e intercambia elementos consecutivos usando punteros.
3. Si no hubo intercambios en una pasada, el algoritmo termina antes.


#### **Pseudocódigo**
```plaintext
Función ordenarBurbuja(punteroArreglo, n)
    Definir bandera ← VERDADERO
    Mientras bandera hacer
        bandera ← FALSO
        Para i ← 0 hasta n - 2 hacer
            Si *(punteroArreglo + i) > *(punteroArreglo + i + 1) entonces
                Intercambiar *(punteroArreglo + i) y *(punteroArreglo + i + 1)
                bandera ← VERDADERO
            FinSi
        FinPara
    FinMientras
FinFunción
```
#### **Explicación del Pseudocódigo**
1. **Uso de punteros**:
   - En lugar de `A[i]`, usamos `*(puntero + i)`, lo que permite manipular directamente la memoria.
2. **Intercambio de valores**:
   - Se intercambian los valores **sin usar índices**, solo con punteros.
3. **Optimización con `bandera`**:
   - Si en una iteración no hubo intercambios, significa que el arreglo ya está ordenado y **el algoritmo se detiene antes de tiempo**.

#### **Ejemplo Paso a Paso con Punteros**
Dado el arreglo `A = [4, 2, 7, 1]`:

##### **Iteraciones del algoritmo**
1. **Primera pasada** (compara e intercambia usando punteros)
   ```
   [4, 2, 7, 1] → [2, 4, 7, 1] → [2, 4, 1, 7]
   ```
2. **Segunda pasada**
   ```
   [2, 4, 1, 7] → [2, 1, 4, 7]
   ```
3. **Tercera pasada**
   ```
   [1, 2, 4, 7]  (ordenado)
   ```

#### **Problema**

1. Solicite un arreglo de enteros.
2. Ordene el arreglo **usando punteros** en lugar de índices.

#### **Objetivo**

### **9. Calculadora con Punteros a Funciones**
#### **Introducción**
Las **funciones en C** pueden ser apuntadas por variables, permitiendo ejecutar código de forma dinámica. Una **calculadora** es un programa que permite realizar operaciones matemáticas básicas como **suma, resta, multiplicación y división**. En C, podemos implementar una calculadora modular y flexible usando **punteros a funciones**.

Un **puntero a función** es una variable que almacena la dirección de una función y permite **llamarla dinámicamente**, sin necesidad de usar estructuras `switch` o múltiples `if-else`.

Esto permite:
- **Seleccionar una operación en tiempo de ejecución**.
- **Reducir el código repetitivo**.
- **Aumentar la eficiencia y flexibilidad** en el programa.

#### **Explicación Matemática**
Dada una operación matemática $ f(x, y) $, donde $ x $ y $ y $ son los operandos y $ f $ es una función matemática, tenemos:

| Operación | Expresión matemática | Función en C |
|-----------|----------------------|--------------|
| Suma      | $ f(x, y) = x + y $ | `sumar(x, y)` |
| Resta     | $ f(x, y) = x - y $ | `restar(x, y)` |
| Multiplicación | $ f(x, y) = x \times y $ | `multiplicar(x, y)` |
| División  | $ f(x, y) = x / y $, $ y \neq 0 $ | `dividir(x, y)` |

Cada operación es representada por una **función** en C, y se asocia dinámicamente a través de un **puntero a función**.

#### **Estrategia Algorítmica**
1. **Definir las funciones** para cada operación matemática.
2. **Declarar un puntero a función** que pueda almacenar la dirección de cualquiera de estas funciones.
3. **Solicitar al usuario que elija una operación** (`+`, `-`, `*`, `/`).
4. **Asignar la función correcta al puntero** según la selección del usuario.
5. **Llamar a la función a través del puntero** con los valores ingresados.

#### **Pseudocódigo**
```plaintext
Definir función sumar(a, b) → Retorna a + b
Definir función restar(a, b) → Retorna a - b
Definir función multiplicar(a, b) → Retorna a * b
Definir función dividir(a, b)
    Si b ≠ 0 entonces
        Retorna a / b
    Sino
        Escribir "Error: División por cero"
        Retorna NULL
    FinSi
FinFunción

Función calcular(operador, x, y)
    Definir puntero a función operacion
    Si operador = '+' entonces
        operacion ← dirección de función sumar
    Sino si operador = '-' entonces
        operacion ← dirección de función restar
    Sino si operador = '*' entonces
        operacion ← dirección de función multiplicar
    Sino si operador = '/' entonces
        operacion ← dirección de función dividir
    Sino
        Escribir "Operador no válido"
        Retornar
    FinSi
    resultado ← llamar a función operacion con parámetros x, y
    Escribir "Resultado: ", resultado
FinFunción

Inicio
    Leer x, y
    Leer operador
    calcular(operador, x, y)
Fin
```

#### **Explicación del Pseudocódigo**
1. **Definir funciones**:
   - `sumar()`, `restar()`, `multiplicar()`, `dividir()`.
   - `dividir()` maneja el caso especial de **división por cero**.
2. **Uso de punteros a funciones**:
   - Un puntero `operacion` almacena la dirección de la función correspondiente.
3. **Asignación dinámica**:
   - Según el operador (`+`, `-`, `*`, `/`), el puntero `operacion` apunta a la función correcta.
4. **Llamado a la función**:
   - Se usa `resultado = operacion(x, y)` para calcular dinámicamente.

#### **Ejemplo de Ejecución**
##### **Entrada**
```
Ingrese el primer número: 10
Ingrese el operador (+, -, *, /): *
Ingrese el segundo número: 5
```

##### **Ejecución**
- `operacion` apunta a `multiplicar()`.
- Se ejecuta `multiplicar(10, 5)`.
- Se imprime `Resultado: 50`.

##### **Salida**
```
Resultado: 50
```

### **10. Implementación de una Pila (Uso Avanzado de Punteros)**
#### **Introducción**
Una **pila (stack)** es una estructura LIFO (último en entrar, primero en salir).

En **C**, las pilas se pueden implementar de dos formas:
1. **Usando arreglos** (tamaño fijo, estático).
2. **Usando listas enlazadas con punteros** (dinámico, sin límite fijo).

Para este ejercicio, implementaremos la pila con **memoria dinámica y punteros**, permitiendo insertar y eliminar elementos sin restricciones de tamaño.

#### **Concepto Matemático**
Una **pila** soporta tres operaciones fundamentales:

1. **Push (Apilar)**: Agrega un nuevo elemento en la parte superior de la pila.
2. **Pop (Desapilar)**: Elimina y devuelve el último elemento agregado.
3. **Mostrar (Peek/Print)**: Imprime el estado actual de la pila.

Si la pila es vacía (`NULL`), no se pueden realizar operaciones `pop()`.

#### **Representación de la Pila**
Cada **nodo** de la pila contiene:
- **Un dato** (`valor`).
- **Un puntero al siguiente nodo** (`siguiente`).

La pila se representa visualmente así:

```
[ TOPE ] → [ Dato: 5 | * ] → [ Dato: 10 | * ] → [ Dato: 15 | NULL ]
```
Donde `*` es un puntero al siguiente nodo.

#### **Estrategia Algorítmica**
1. **Crear la estructura del nodo** con un `struct`.
2. **Implementar `push()`** para agregar elementos a la pila.
3. **Implementar `pop()`** para eliminar el último elemento ingresado.
4. **Implementar `mostrar()`** para imprimir la pila.
5. **Liberar memoria** al desapilar (`free()`).

#### **Pseudocódigo**
##### **Definir la estructura del nodo**
```plaintext
Definir estructura Nodo:
    Dato → entero
    Siguiente → Puntero a Nodo
```

##### **Operación `push(valor)` - Insertar un elemento**
```plaintext
Función push(Puntero &pila, valor)
    Crear nuevoNodo con valor
    nuevoNodo.Siguiente ← pila  // Apunta al nodo anterior
    pila ← nuevoNodo  // Ahora el nuevo nodo es el tope de la pila
FinFunción
```

**Ejemplo de `push(5)`:**
```
Antes: NULL
Después: [5 | NULL]
```

Si luego hacemos `push(10)`, la pila queda así:
```
[10 | *] → [5 | NULL]
```

##### **Operación `pop()` - Eliminar un elemento**
```plaintext
Función pop(Puntero &pila)
    Si pila es NULL entonces
        Escribir "Error: Pila vacía"
        Retornar NULL
    FinSi
    temp ← pila
    pila ← pila.Siguiente  // Mover tope de la pila
    valor ← temp.Dato  // Guardar valor antes de eliminar
    Liberar memoria temp
    Retornar valor
FinFunción
```

**Ejemplo de `pop()` en la pila `[10 | *] → [5 | NULL]`**
1. `pop()` devuelve `10` y elimina el nodo `10`.
2. La pila queda `[5 | NULL]`.

##### **Operación `mostrar()` - Imprimir la pila**
```plaintext
Función mostrar(Puntero pila)
    Si pila es NULL entonces
        Escribir "Pila vacía"
        Retornar
    FinSi
    temp ← pila
    Mientras temp ≠ NULL hacer
        Escribir temp.Dato
        temp ← temp.Siguiente
    FinMientras
FinFunción
```

Ejemplo de `mostrar()` para la pila `[15 | *] → [10 | *] → [5 | NULL]`:

```
15
10
5
```

#### **Ejemplo Paso a Paso**
##### **Insertar elementos (`push`)**
1. `push(5) → [5 | NULL]`
2. `push(10) → [10 | *] → [5 | NULL]`
3. `push(15) → [15 | *] → [10 | *] → [5 | NULL]`

##### **Eliminar elemento (`pop`)**
1. `pop() → Devuelve 15 y la pila queda [10 | *] → [5 | NULL]`
2. `pop() → Devuelve 10 y la pila queda [5 | NULL]`
3. `pop() → Devuelve 5 y la pila queda `NULL` (vacía).

##### **Imprimir (`mostrar()`)**
Si la pila es `[10 | *] → [5 | NULL]`, la salida será:

```
10
5
```

#### **Complejidad Temporal**
| Operación | Complejidad |
|-----------|------------|
| **Push (Insertar)** | O(1) |
| **Pop (Eliminar)** | O(1) |
| **Mostrar** | O(n) |

- **Push y Pop son O(1)** porque solo afectan el tope de la pila.
- **Mostrar es O(n)** porque recorre toda la pila.

#### **Resumen**
✔ **Estructura LIFO**: Último en entrar, primero en salir.  
✔ **Uso de memoria dinámica**: Permite tamaño flexible.  
✔ **Eficiencia O(1) en `push()` y `pop()`**.  
✔ **Aplicaciones**: Algoritmos de retroceso (*backtracking*), análisis de expresiones matemáticas, sistemas de control de llamadas, etc.
