Escribe un programa que paralelice el cálculo del producto punto entre dos vectores de tamaño 10 millones de elementos.

```c
// Pseudocódigo
- Inicializar dos vectores con números aleatorios.
- Implementar el cálculo del producto punto en un bucle simple.
- Utilizar `#pragma omp simd` para vectorizar el cálculo.
```

**Desafío adicional**: Implementa la misma operación utilizando `reduction` y compara la eficiencia de ambas versiones.


Para la solución de este punto se realizará inicialmente un código secuencial para compararla con nuestras versiones paralelizadas

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define N 10000000

int main() {
    float *a = malloc(sizeof(float) * N);
    float *b = malloc(sizeof(float) * N);
    float dot = 0.0;

    for (int i = 0; i < N; i++) {
        a[i] = (float)rand() / RAND_MAX;
        b[i] = (float)rand() / RAND_MAX;
    }

    clock_t start = clock();

    for (int i = 0; i < N; i++) {
        dot += a[i] * b[i];
    }

    clock_t end = clock();
    printf("Dot product: %f\n", dot);
    printf("Time (sequential): %f seconds\n", (double)(end - start) / CLOCKS_PER_SEC);

    free(a); free(b);
    return 0;
}
```
